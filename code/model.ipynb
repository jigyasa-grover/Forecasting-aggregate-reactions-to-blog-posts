```python
df = pd.read_csv("../dataset/blogData_train.csv", header=None, names=headers, index_col=False)
```

```python
df = df[df.length_of_blog_post > 100]
```

```python
# Log transforming the data
df['target_number_of_comments_in_next_24_hours'] = np.log(1 + df['target_number_of_comments_in_next_24_hours'])
```

```python
df.describe()
```

```python
array = df.values
X = array[:, 0:280]
y = array[:, 280]
```

```python
model = PCA(n_components=50)
model.fit(X, y)

X = model.transform(X)
```

```python
def error(y_true, y_pred):
    a = ((np.log(1 + np.array(y_true)) - np.array(y_pred)) ** 2).sum()
    return a / len(y_true)
```

```python
scoring = make_scorer(error, greater_is_better=False)
```

```python
seed = 7
kfold = model_selection.KFold(n_splits=10, random_state=seed)
#scoring = 'neg_mean_squared_error'
```

```python
# SVM Regression
model = SVR()
results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
print(results.mean())
```

```python
model = ExtraTreesRegressor()
model.fit(X, y)

imp = model.feature_importances_
indices = np.argpartition(imp, -50,)[-50:]

X = df.iloc[:, indices].values
```

```python
# Lasso Regression
model = Lasso()
results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
print(results.mean())

# KNN Regression
model = KNeighborsRegressor()
results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
print(results.mean())

# DecisionTree Regression
model = DecisionTreeRegressor()
results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
print(results.mean())

# SVM Regression
#model = SVR()
#results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
#print(results.mean())



# MLP Regression
model = MLPRegressor()
results = model_selection.cross_val_score(model, X, y, cv=kfold, scoring=scoring)
print(results.mean())
```
